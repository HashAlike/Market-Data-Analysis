```python
#Imported Python libraries

import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns
import yfinance as yf 
from datetime import date, timedelta
```

```python
#Specify the date range.
end_date = date.today()
start_date=end_date - timedelta(days=5*365) # 5 years ago
```

```python
TICKER_LIST = [
    'AAPL', 'MSFT', 'NVDA', 'TSLA', 'XOM', 
    'JPM', 'BTC-USD', '^GSPC', '^IXIC', 'GLD'
]                                                       #list of tickers to download
```

```python
# Writing the Data Retrieval Function
from yfinance import tickers


def fetch_data(ticker_list,start_date,end_date):
    # Retrieves price data for the specified symbols.
    data= yf.download(
        tickers=ticker_list,
        start=start_date,
        end=end_date,
        interval="1d"
    )
    return data
financial_data= fetch_data(TICKER_LIST, start_date, end_date)
```

```python
# Metric filtering step.
close_prices= financial_data['Close']  # Extracting the 'Close' prices from the data
print(close_prices.head())  # Display the first few rows of the close prices
```

```python
close_prices.columns.name= None # Remove the name of the columns index for cleaner display
close_prices.columns.name= 'Date' # Set the index name as 'Date.'

close_prices.info() # Display information about the close prices DataFrame
```

```python
close_prices.isnull() # Check for missing values in the close prices DataFrame
```

```python
close_prices.isnull().sum() # Count the number of missing values in each column of the close prices DataFrame
```

```python
close_prices.isnull().values.any() # Check if there are any missing values in the close prices DataFrame
```

```python
close_prices.dropna(inplace=True) # Drop rows with missing values
close_prices.dropna(axis=1) # Drop columns with missing values
close_prices.info() # Display information about the close prices DataFrame after handling missing values
```

```python
df_filled= close_prices.fillna("Unknown") # Fill missing values with a specified value
df_filled.info() # Display information about the DataFrame after filling missing values
print(df_filled.head())  # Display the first few rows of the filled DataFrame
```
