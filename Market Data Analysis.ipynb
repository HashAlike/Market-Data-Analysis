```python
# Writing the Data Retrieval Function
from yfinance import tickers


def fetch_data(ticker_list,start_date,end_date):
    # Retrieves price data for the specified symbols.
    data= yf.download(
        tickers=ticker_list,
        start=start_date,
        end=end_date,
        interval="1d"
    )
    return data
financial_data= fetch_data(TICKER_LIST, start_date, end_date)
```
